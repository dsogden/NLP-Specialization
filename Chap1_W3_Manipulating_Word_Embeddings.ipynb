```python
import pandas as pd
import numpy as np
import pickle

word_embeddings = pickle.load(open('/content/drive/MyDrive/word_embeddings_subset.p', 'rb'))
len(word_embeddings)
```

```python
country_vector = word_embeddings['country']
print(type(country_vector))
print(country_vector)
```

```python
def vec(word_embed, w):
    return word_embed[w]
```

```python
import matplotlib.pyplot as plt

words = [
    'oil', 'gas', 'happy', 'sad', 'city',
    'town', 'village', 'country', 'continent',
    'petroleum', 'joyful'
]

bag2d = np.array([vec(word_embeddings, word) for word in words])

fig, ax = plt.subplots(figsize=(10, 8))
col1 = 3
col2 = 2

for word in bag2d:
    ax.arrow(
        0, 0, word[col1], word[col2], head_width=0.005, head_length=0.005,
        fc='r', ec='r', width=1E-5
    )
ax.scatter(bag2d[:, col1], bag2d[:, col2])

for i in range(len(words)):
    ax.annotate(words[i], (bag2d[i, col1], bag2d[i, col2]))

ax.set(
    xlim=[-0.3, 0.3],
    ylim=[-0.3, 0.3],
    xlabel='$Vec_{X}$',
    ylabel='$Vec_{Y}$'
)
plt.show()
```

## Word Distance

```python
words = ['sad', 'happy', 'town', 'village']
bag2d = np.array([vec(word_embeddings, word) for word in words])

fig, ax = plt.subplots(figsize=(10, 8))
col1 = 3
col2 = 2

for word in bag2d:
    ax.arrow(
        0, 0, word[col1], word[col2], head_width=0.005, head_length=0.005,
        fc='r', ec='r', width=1E-5
    )

village = vec(word_embeddings, 'village')
town = vec(word_embeddings, 'town')
diff = town - village
ax.arrow(
    village[col1], village[col2], diff[col1], diff[col2], fc='b',
    ec='b', width=1E-5
)

sad = vec(word_embeddings, 'sad')
happy = vec(word_embeddings, 'happy')
diff = happy - sad
ax.arrow(
    sad[col1], sad[col2], diff[col1], diff[col2], fc='b',
    ec='b', width=1E-5
)

ax.scatter(bag2d[:, col1], bag2d[:, col2])

for i in range(len(words)):
    ax.annotate(words[i], (bag2d[i, col1], bag2d[i, col2]))

ax.set(
    xlim=[-0.25, 0.1],
    ylim=[-0.05, 0.15],
    xlabel='$Vec_{X}$',
    ylabel='$Vec_{Y}$'
)
plt.show()
```

##Linear algebra on word embeddings

```python
town = np.linalg.norm(vec(word_embeddings, 'town'))
sad = np.linalg.norm(vec(word_embeddings, 'sad'))

print(f'Town norm: {town}')
print(f'Sad norm: {sad}')
```

## Predicting capitals

```python
capital = vec(word_embeddings, 'France') - vec(word_embeddings, 'Paris')
country = vec(word_embeddings, 'Madrid') + capital
print(country[:5])
```

```python
diff = country - vec(word_embeddings, 'Spain')
print(diff[:10])
```

```python
keys = word_embeddings.keys()
data = []
for key in keys:
    data.append(word_embeddings[key])
embedding = pd.DataFrame(data=data, index=keys)

def find_closest_word(v, k=1):
    diff = embedding.values - v
    delta = np.sum(diff ** 2, axis=1)
    i = np.argmin(delta)
    return embedding.iloc[i].name
```

```python
find_closest_word(country)
```

## Predicting other countries

```python
print(find_closest_word(vec(word_embeddings, 'Berlin') + capital))
print(find_closest_word(vec(word_embeddings, 'Beijing') + capital))
```

```python
print(find_closest_word(vec(word_embeddings, 'Lisbon') + capital))
```

## Represent a sentence as a vector

```python
doc = 'Spain petroleum city king'
vdoc = [vec(word_embeddings, x) for x in doc.split(' ')]
doc2vec = np.sum(vdoc, axis=0)
doc2vec
```

```python
find_closest_word(doc2vec)
```
