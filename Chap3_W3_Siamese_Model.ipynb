```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.models import Model, Sequential
from tensorflow import math
import numpy as np

np.random.seed(10)
```

```python
vocab_size = 500
hidden_size = 128

sequential = Sequential([
    layers.Embedding(input_dim=vocab_size, output_dim=hidden_size),
    layers.LSTM(units=hidden_size, return_sequences=True),
    layers.AveragePooling1D(),
    layers.Lambda(lambda x: math.l2_normalize(x))
])

inputs_1 = layers.Input((None, ))
inputs_2 = layers.Input((None, ))

concatenate = layers.Concatenate(axis=1)((sequential(inputs_1), sequential(inputs_2)))
model = Model(inputs=(inputs_1, inputs_2), outputs=concatenate)
model.summary()
```

```python
rands1 = np.random.rand(10, 10)
rands2 = np.random.rand(10, 10)

model((rands1, rands2))
```
